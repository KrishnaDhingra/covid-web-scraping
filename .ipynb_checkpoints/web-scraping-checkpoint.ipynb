```python
from bs4 import BeautifulSoup
import requests
```

```python
def getPageHTML(url):
    result = requests.get(url)
    doc = BeautifulSoup(result.text, 'html.parser')
    return doc

print(getPageHTML('https://www.worldometers.info/coronavirus/'))
```
