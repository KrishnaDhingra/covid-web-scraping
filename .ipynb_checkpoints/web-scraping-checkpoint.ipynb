```python
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import requests
```

```python
def getPageHTML(url):
    result = requests.get(url)
    doc = BeautifulSoup(result.text, 'html.parser')
    return doc
```

```python
def getGlobalInfo(doc):
    
    tableBody = doc.find('table', {"id": "main_table_countries_today"}).tbody
    rows = tableBody.find_all("tr", style="")[1:]

    globalInfo = []
    for row in rows:
        countryInfo = {}
        tableData = row.find_all('td')

        countryInfo["country"] = tableData[1].a.string
        countryInfo["total cases"] = tableData[2].string
        countryInfo["total deaths"] = tableData[4].string
        countryInfo["total recovered"] = tableData[6].string
        countryInfo["active cases"] = tableData[8].string
        countryInfo["serious critical"] = tableData[9].string
        countryInfo["total tests"] = tableData[12].string
        
        if tableData[14].a !=None:
            countryInfo["population"] = tableData[14].a.string
        else:
            countryInfo["population"] = tableData[14].string
        globalInfo.append(countryInfo)
        
    return globalInfo
```

```python
def getCountry(userCountry, globalInfo):
    for countryInfo in globalInfo: 
        if countryInfo['country'].lower() == userCountry.lower():
            return True, countryInfo
    return False, "Can't find the country"
```

```python
def plotGraph(countryInfo):
    x = list(countryInfo.keys())
    h = list(countryInfo.values())
    plt.bar(x, h)
    plt.title('Covid information')
    plt.show()
```

```python
def askForSpecificCountry(globalInfo):
    while True:
        userCountry = input("Please type the country of whose information you want to get or type 'skip' to move forward: ")
        if userCountry.lower() != 'skip':
            exists, message = getCountry(userCountry, globalInfo)
#           if the getCountry function returns true than the country exists the the country
            if(exists):
#               print an empty line so it looks good
                print()
#               using this loop the country is printed in a pretty manner
                for i in message:
                    print(f'{i} : {message[i]}')
                print()
                plotGraph(message)
            else:
#               if the country doesn't exists the just print the message
                print()
                print(message)
        else: 
            break
```

```python
doc = getPageHTML('https://www.worldometers.info/coronavirus/')
globalInfo = getGlobalInfo(doc)
    
for country in globalInfo:
    for countryInfo in country:
        print(f'{countryInfo} : {country[countryInfo]}')
    print()
    
askForSpecificCountry(globalInfo)
```
